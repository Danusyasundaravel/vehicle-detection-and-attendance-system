```python
from ultralytics import YOLO
import cv2
from paddleocr import PaddleOCR
import pandas as pd
import time
import os

# Load the YOLO model
model = YOLO('/content/custom_model.pt')  # Replace with your model file

# Initialize PaddleOCR
ocr = PaddleOCR(use_angle_cls=True, lang='en')  # OCR for detecting text

# Path for the CSV file where the data will be stored
csv_filename = '/content/number_plate_texts.csv'

# Function to read the current CSV file and check if a number plate is already saved
def read_existing_data(csv_filename):
    if os.path.exists(csv_filename):
        return pd.read_csv(csv_filename)
    else:
        # Return an empty DataFrame with columns if file does not exist
        return pd.DataFrame(columns=['Number Plate Text', 'Confidence'])

# Function to detect and extract number plates using YOLO and PaddleOCR
def detect_and_extract_number_plate(image_path):
    # Read the image
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Perform detection using YOLO
    results = model.predict(image_rgb, conf=0.5)

    # Store detected number plates' text and confidence
    extracted_texts = []

    # Iterate over each detection result
    for result in results:
        boxes = result.boxes  # Bounding boxes for detected objects
        for box in boxes:
            # Get coordinates of the detected number plate
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Coordinates of the bounding box
            cropped_image = image[y1:y2, x1:x2]  # Crop the number plate region

            # Use PaddleOCR to extract text from the cropped image
            ocr_result = ocr.ocr(cropped_image, cls=True)

            # Extract text from OCR results
            if ocr_result:
                plate_text = ocr_result[0][0][1]  # Extract the text (first element in result)
                confidence = box.conf[0]  # Extract the confidence score of the detected plate
                extracted_texts.append([plate_text, confidence])

    return extracted_texts

# Function to save new detected number plates and append to the CSV
def save_to_csv(data, csv_filename):

    # Read existing data from CSV file, creating a new one with correct columns if it doesn't exist
    if os.path.exists(csv_filename):
        existing_data = pd.read_csv(csv_filename)
    else:
        existing_data = pd.DataFrame(columns=['Number Plate Text', 'Confidence'])  # Define columns here

    # Convert extracted data to DataFrame
    new_data = pd.DataFrame(data, columns=['Number Plate Text', 'Confidence'])

    # Check for duplicates and append only new entries
    for index, row in new_data.iterrows():
        # Check if the 'Number Plate Text' is already in the existing data
        if row['Number Plate Text'] not in existing_data['Number Plate Text'].to_list():  # Convert to list for comparison
            # Append new rows if number plate is not already in the CSV
            existing_data = pd.concat([existing_data, row.to_frame().T], ignore_index=True)

    # Save the updated DataFrame to CSV
    existing_data.to_csv(csv_filename, index=False)
    print(f"Data saved to {csv_filename}")

if __name__ == "__main__":
    image_path = "/content/8egca.png"  # Replace with your image path

    # Step 1: Detect number plates and extract text
    extracted_texts = detect_and_extract_number_plate(image_path)

    # Step 2: Save the extracted texts to CSV
    if extracted_texts:
        save_to_csv(extracted_texts, csv_filename)
    else:
        print("No number plates detected.")
```

```python
from ultralytics import YOLO
from paddleocr import PaddleOCR
import cv2
import pandas as pd
import time

# Load YOLO model (replace with your trained model)
model = YOLO('custom_model.pt')  # Replace with your model file

# Initialize PaddleOCR for text extraction
ocr = PaddleOCR(use_angle_cls=True, lang='en')

# Path to store attendance data
attendance_file = '/content/number_plate_texts.csv'

# Function to read attendance data from CSV (if it exists)
def read_attendance():
    if pd.io.common.file_exists(attendance_file):
        return pd.read_csv(attendance_file)
    else:
        # Create an empty dataframe if no records exist
        return pd.DataFrame(columns=['Number Plate Text ', 'Date', 'Time'])

def mark_attendance(number_plate):
    # Get current date and time
    current_time = time.strftime("%H:%M:%S")
    current_date = time.strftime("%Y-%m-%d")

    # Read existing attendance records
    attendance_df = read_attendance()

    # Check if number plate already exists
    if number_plate in attendance_df['Number Plate Text'].values.tolist():  # Move to CPU and convert to list
        print(f"Attendance for {number_plate} already marked.")
    else:
        # Mark attendance
        new_entry = pd.DataFrame([[number_plate, current_date, current_time]], columns=['Number Plate Text', 'Date', 'Time'])
        attendance_df = pd.concat([attendance_df, new_entry], ignore_index=True)

        # Save the updated attendance to the CSV file
        attendance_df.to_csv(attendance_file, index=False)
        print(f"Attendance marked for {number_plate} at {current_time} on {current_date}")

# Function to detect number plates and extract text
def detect_and_extract_number_plate(image_path):

    # Read the image
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Perform detection using YOLO
    results = model.predict(image_rgb, conf=0.5)

    # Extracted text list
    extracted_texts = []

    # Iterate over detection results
    for result in results:
        boxes = result.boxes  # Bounding boxes for detected objects
        for box in boxes:
            # Get coordinates of the detected number plate
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Coordinates of the bounding box
            cropped_image = image[y1:y2, x1:x2]  # Crop the number plate region

            # Use PaddleOCR to extract text from the cropped image
            ocr_result = ocr.ocr(cropped_image, cls=True)
            if ocr_result:
                plate_text = ocr_result[0][0][1]  # Extract the text (first element in result)
                extracted_texts.append(plate_text)

    # Return the extracted number plate text (if detected)
    return extracted_texts[0] if extracted_texts else None

# Function to handle the upload and attendance marking
def handle_attendance(image_path):
    # Extract the number plate text
    number_plate = detect_and_extract_number_plate(image_path)

    if number_plate:
        print(f"Detected Number Plate: {number_plate}")

        # Mark attendance if number plate is found
        mark_attendance(number_plate)
    else:
        print("No number plate detected in the image.")

# Example usage
if __name__ == "__main__":
    # Upload or provide the path to your test image
    from google.colab import files
    uploaded = files.upload()

    # Get the uploaded image path (as a string)
    image_path = list(uploaded.keys())[0]
    # Replace with your image path

    # Handle attendance marking
    handle_attendance(image_path)
```
